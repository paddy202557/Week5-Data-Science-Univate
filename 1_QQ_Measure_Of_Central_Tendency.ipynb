# Central Tendency

```python
dataset["ssc_p"]
```

```python
#The mean() function -It adds up all the values in a list and divides the total by the number of elements.
#This function is useful when analyzing data and finding central tendencies.
dataset["ssc_p"].mean()
```

```python
#median() method calculates the median (middle value) of the given data set. 
#This method also sorts the data in ascending order before calculating the median.
#Median = {(n + 1) / 2}th value, where n is the number of values in a set of data. 
#In order to calculate the median, the data must first be sorted in ascending order. 
#The median is the number in the middle.
dataset["ssc_p"].median()
```

```python
#The mode() is a measure of central tendency that represents the most common value in a dataset.
#A dataset can have: a) No mode: If all values are unique b) One mode: If one value appears more frequently than others (unimodal)
# c) Multiple modes: If two or more values appear with the same highest frequency (bimodal or multimodal).
dataset["ssc_p"].mode()[0]
```

```python
#The describe() method returns description of the data in the DataFrame.
#If the DataFrame contains numerical data, the description contains these information for each column:(Pplease See the result)
#Percentile (25%,50%,75%) meaning: how many of the values are less than the given percentile.
dataset.describe()
```

```python
#Assigning the value to mean  index row  from the dataset-quantative columns(Quan1)
descriptive=pd.DataFrame(index=["Mean", "Median", "Mode"],columns=Quan1)
for columnName in Quan1:
    descriptive[columnName]["Mean"]=dataset[columnName].mean()
```

```python
descriptive
```

```python
#Assigning the value to mean, median, mode  index rows  from the dataset-quantative columns(Quan1)
descriptive=pd.DataFrame(index=["Mean", "Median", "Mode"],columns=Quan1)
for columnName in Quan1:
    descriptive[columnName]["Mean"]=dataset[columnName].mean()
    descriptive[columnName]["Median"]=dataset[columnName].median()
    descriptive[columnName]["Mode"]=dataset[columnName].mode()
    
```

```python
descriptive
```

```python
#Assigning the value to Mean, Median, Mode  index rows  from the dataset-quantative columns(Quan1)
#Rounded the value Mean and Mode
descriptive=pd.DataFrame(index=["Mean", "Median", "Mode"],columns=Quan1)
for columnName in Quan1:
    descriptive[columnName]["Mean"]=round(dataset[columnName].mean())
    descriptive[columnName]["Median"]=dataset[columnName].median()
    descriptive[columnName]["Mode"]=dataset[columnName].mode()[0]
```
