# Scenario 5 - Simple Imputer - Replacing the null values in columns  with Central Tendency values

```python
from QQUnivariate import QQUnivariate
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
#SimpleImputer is a class in Scikit-Learn that provides basic strategies for imputing missing values in datasets. 
#It replaces missing values using a specified strategy, such as the mean, median, most frequent value,or a constant value.
```

```python
#Scenario- 5 Replace the Null column value with Mean value
dataset=pd.read_csv("Placement.csv")
```

```python
dataset
```

```python
dataset.describe()
```

```python
Quan1,Qual1=QQUnivariate.quanQual(dataset)
```

```python
Quan1
```

```python
Qual1
```

```python
df_Qual=dataset[Qual1]
```

```python
df_Qual
```

```python
#SimpleImputer is a class in Scikit-Learn that provides basic strategies for imputing missing values in datasets. 
#It replaces missing values using a specified strategy, such as the mean, median, most frequent value,or a constant value.
#SimpleImputer performs univariate imputation, meaning it imputes missing values in each feature independently using only the non-missing values in that feature
#missing_values: The placeholder for the missing values, default is np.nan
```

```python
imp= SimpleImputer(missing_values=np.nan,strategy='mean')
imp.fit(dataset[Quan1])
df=(imp.transform(dataset[Quan1]))
print(imp.transform(dataset[Quan1]))
```

```python
df_Quan =pd.DataFrame(df,columns= Quan1)
```

```python
df_Quan
```

```python
df_Quan.isnull().sum()
```

```python
Qual=dataset[Qual1]
```

```python
Qual
```

```python
Qual.isnull().sum()
```

# Merge Quantitative and Qualitative Data Sets

```python
df_Quan_Qual_merge=[df_Quan,Qual]
```

```python
df_Quan_Qual_merge
```

```python
#axis= 1 parameter  merges columns from two source into a single  record
preprocessed=pd.concat(df_Quan_Qual_merge,axis=1)
```
