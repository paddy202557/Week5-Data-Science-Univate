```python
import pandas as pd
from QQUnivariate import QQUnivariate
import numpy as np
from sklearn.impute import SimpleImputer
```

# Kidney CKD data  from Kaggle
# Replacing the  Null values in  Qualitative  & Qunatative Data

```python
dataset=pd.read_csv("kidneydisease.csv")
```

```python
dataset
```

```python
dataset.isnull().sum()
```

# Identifying the  Qualitative and Quantitative Columns

```python
Quan,Qual=QQUnivariate.quanQual(dataset)
```

```python
Quan
```

```python
Qual
```

```python
dataset.describe()
```

# Replacing the Quantitative Null value Columns  with Mean value

```python
imp= SimpleImputer(missing_values=np.nan,strategy='mean')
imp.fit(dataset[Quan])
df1=(imp.transform(dataset[Quan]))
print(imp.transform(dataset[Quan]))
```

```python
df_Quan =pd.DataFrame(df1,columns= Quan)
```

```python
df_Quan
```

```python
df_Quan.isnull().sum()
```

# Replacing the Qualitative Null value Columns  with Most Frequent Values

```python
#strategy : The data which will replace the NaN values from the dataset. 
#The strategy argument can take the values – ‘mean'(default), ‘median’, ‘most_frequent’ and ‘constant’.
```

```python
imp= SimpleImputer(missing_values=np.nan,strategy='most_frequent')
imp.fit(dataset[Qual1])
df2=(imp.transform(dataset[Qual1]))
print(imp.transform(dataset[Qual1]))
```

```python
df_Qual =pd.DataFrame(df2,columns= Qual)
```

```python
df_Qual
```

```python
df_Qual.isnull().sum()
```

# Merge Quantitative and Qualitative Data Sets

```python
df_Quan_Qual_merge=[df_Quan,df_Qual]
```

```python
df_Quan_Qual_merge
```

```python
#axis= 1 parameter  merges columns from two source into a single  record
preprocessed=pd.concat(df_Quan_Qual_merge,axis=1)
```

```python
preprocessed
```

```python
preprocessed.isnull().sum()
```

# Create the CSV file

```python
#Index = False parameter prevents  duplicate records or index
preprocessed.to_csv("Prekidneydisease.csv",index=False)
```
