# Variance  &  Standard Deviation

```python
# Mean - Average value (Center of the data) Mean = X
# Variance (Squared) help us to understand  the variance is how for from Mean.   
# Population Variance(all data in the data set): Averaging the Variance (Squared value) 
#                                 = Data point value  - Mean value / Number of Data Points
# Sample Varaince (Sample data in  the data set): Averaging the Variance (Squared value) 
#                                = Data point value  - Mean value / Number of Data Points - 1
```

```python
# Standard Deviation (Not Squared) help us to Understand the  spread of the data 
#      standard deviation is often preferred because it is expressed in the same units as the original data
```
